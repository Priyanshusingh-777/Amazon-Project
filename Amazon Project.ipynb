# What is the distribution of Star Ratings?

```python
plt.figure(figsize=(12,6))

sns.histplot(x="stars", data=df, bins=20, kde=True)
plt.title("Distribution of Stars Ratings", fontsize=16)
plt.xlabel("Stars Ratings", fontsize=16)
plt.ylabel("Frequency", fontsize=16)
plt.show()
```

# Number of Review vs Price

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x="price", y="reviews", data=df, hue="isBestSeller", palette='coolwarm')
plt.title("Price vs Number of Reviews", fontsize=16)
plt.xlabel("Price", fontsize=16)
plt.ylabel("Number of Reviews", fontsize=16)
plt.legend(title="Best Seller", loc="upper right", fontsize=10)
plt.show()
```

# Price distribution for best sellers

```python
import warnings
warnings.filterwarnings("ignore")
plt.figure(figsize=(10,6))
sns.boxplot(x="isBestSeller", y="price", data=df, palette='Set3')
plt.title("Price Distribution for Best Sellers", fontsize=16)
plt.xlabel("Is Best Seller", fontsize=16)
plt.ylabel("Price", fontsize=16)
plt.xticks([0,1],["Not Best Seller", "Best Seller"], fontsize=14)
plt.show()
```

# How does the number of reviews vary with star ratings?

```python
plt.figure(figsize=(16,9))
sns.violinplot(x="stars", y="reviews", data=df, palette='viridis')
plt.title("Distribution of Reviews by Star Ratings", fontsize=16)
plt.xlabel("Star Ratings", fontsize=14)
plt.ylabel("Number of Reviews", fontsize=14)
plt.show()
```

# Is there a relationship between price and list price?

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x="price", y="listPrice", data=df, 
                hue='isBestSeller', palette='coolwarm')
plt.title("Price vs List Price", fontsize=16)
plt.xlabel("Price", fontsize=14)
plt.ylabel("List Price", fontsize=14)
plt.legend(title="Best Seller", loc="upper right", fontsize=12)
plt.show()
```

# How many products are best sellers and were bought in the last month?

```python
plt.figure(figsize=(10,6))
sns.countplot(x="isBestSeller", data=df, hue='boughtInLastMonth', palette='coolwarm')
plt.title("Best Seller Bought in the Last Month", fontsize=16)
plt.xlabel("Is Best Seller", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.legend(title="Bought in Last Month", loc="upper right", fontsize=14)
plt.xticks([0,1],["Not Bought", "Bought"], fontsize=12)
plt.show()
```

# Correaltion between customer ratings and the number of reviews:

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="stars", y="reviews", data=df, hue='isBestSeller', palette='tab10')
plt.title("Correlation Between Customer Ratings and Number of Reviews", fontsize=16)
plt.xlabel("Stars",fontsize=16)
plt.ylabel("Number of Reviews",fontsize=16)
plt.show()
```

```python
df['Discount Percentage'] = ((df['listPrice'] - df['price']) / df['listPrice']) * 100
```

```python
df.head(3)
```

```python
df['Discount Percentage'] = df['Discount Percentage'].round(0)
```

```python
df.head(3)
```

```python
from IPython.display import display, HTML
```

# Top 10 Stars on which Product? and how much was the Discount?

```python
top_10_discount_stars = df.sort_values(by="Discount Percentage", ascending=False).head(10)
top_10_discount_stars[['title', 'Discount Percentage', 'stars']]
```

# Top 5 Products with highest discount

```python
top_5_prod = df.nlargest(5, 'Discount Percentage')['imgUrl']

for imgUrl in top_5_prod:
    display(HTML(f'<img src = "{imgUrl}">'))   
```

# Top 5 Products with Highest Ratings

```python
top_5_images = df.nlargest(5, "stars")['imgUrl']

for imgUrl in top_5_images:
    display(HTML(f'<img src = "{imgUrl}">'))
```

# Top 5 Products with Highest Price

```python
top_5_price = df.nlargest(5, "price")["imgUrl"]

for imgUrl in top_5_price:
    display(HTML(f'<img src = "{imgUrl}">'))
```
